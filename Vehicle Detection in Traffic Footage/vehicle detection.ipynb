```python
from ultralytics import YOLO
import cv2

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Open video (upload your own file or use sample.mp4)
cap = cv2.VideoCapture("traffic.avi")

# Prepare output video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("detected_output.mp4", fourcc, 20.0,
                      (int(cap.get(3)), int(cap.get(4))))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference
    results = model(frame)
    annotated = results[0].plot()      # Draw boxes and labels
    out.write(annotated)               # Save annotated frame

cap.release()
out.release()
cv2.destroyAllWindows()
print("✅ Detection video saved as detected_output.mp4")
```
